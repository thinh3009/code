```python
def convert_date_format(dt_str):
    dt = pd.to_datetime(dt_str, format='%d/%m/%Y')
    return dt.strftime('%y/%m/%d')
```

```python
df['Time'] = df['Time'].apply(convert_date_format)
```

```python
df.to_excel('processed_file3.xlsx', index=False)
```

```python
def change_year(dt_str):
    dt = pd.to_datetime(dt_str, format='%y/%m/%d')
    # Thay đổi năm thành 2022
    dt = dt.replace(year=2022)
    return dt.strftime('%y/%m/%d')
```

```python
df['Time'] = df['Time'].apply(change_year)
```

```python
def combine_date_time(date_str, time_str):
    return f"{date_str} {time_str}"
```

```python
df['Combined'] = df.apply(lambda row: combine_date_time(row['Time'], row['Column1']), axis=1)
```

```python
df.to_excel('processed_file5.xlsx', index=False)
```
