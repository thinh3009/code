```python
#tkinter
from tkinter import *
root = Tk()#bat buoc khai bao
root.title("http://hoc.com")
root.resizable(height=True, width=True)#thay doi kich thuoc chieu cao chieu rong
root.minsize(height=300,width=500)#kich thuoc toi thieu cua bang

#dua giao dien vao giua man hinh
def makeCenter(root):
    root.update_idletasks()
    width=root.winfo_width()
    height=root.winfo_height()
    x=(root.winfo_screenwidth()//2)-(width//2)
    y=(root.winfo_screenwidth()//2)-(height//2)
    root.geometry('{}x{}+{}+{}'.format(width,height,x,y))
makeCenter(root)
root.mainloop()

```

```python
#control
#.pack de hien thi len noi dung
from tkinter import  *
root=Tk()
root.title("web")
root.resizable(height=True,width=True)
root.minsize(height=200,width=300)
Label(root,text="hello tkinter",fg="red").pack()#giao dien chu
#Button(root,text="click").pack()#nut bam
Button(root,text="click",command=root.quit).pack()#nut bam de thoat bang
e=StringVar()
e.set('xin chao')#thay doi giao dien khung text
Entry(root,textvariable=e,width=30).pack()#tao khung viet text
root.mainloop()
```
