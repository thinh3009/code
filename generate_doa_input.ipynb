```python

import pandas as pd
import os

```

```python

users_filepath = os.path.join("./data", files[0])
# print(users_filepath)
users_xlx = pd.ExcelFile(users_filepath)


# Initialize an empty list to hold DataFrames
dfs = []

# Loop through each sheet in the Excel file
for sheet_name in users_xlx.sheet_names:
    # Load the sheet into a DataFrame
    r_df = pd.read_excel(users_filepath, sheet_name=sheet_name)
    r_df.rename(columns={'Unnamed: 0': 'Name', 'Unnamed: 1': 'Email'}, inplace=True)
    # Add a new column 'BU' with the value as the sheet name
    r_df['BU'] = sheet_name
    r_df['Email'] = r_df['Email'].str.strip().str.replace(',', '.')
    
    # Append the DataFrame to the list
    dfs.append(r_df)

# Concatenate all DataFrames into one
df = pd.concat(dfs, ignore_index=True)
# Rename columns 'Unnamed: 0' and 'Unnamed: 1' to 'name' and 'email'
df.head(10)
```

```python
# Define a function to determine the correct permission
def map_permission(value):
    if value:
        value_lower = value.strip().lower()
        if value_lower in ['e', 'edit', 'edit-all', 'e-all', 'e - all', 'x']:
            return 'Edit'
        elif value_lower in ['v', 'view', 'view-all', 'v-all']:
            return 'View'
    return value  # Default: keep the original value if not matched
```

```python
# KH PRJ - Extract relevant rows
khprj_df = df.dropna(subset=['Folder Xây dựng KH PRJ']).copy()  # Create a copy to avoid SettingWithCopyWarning

# Rename the relevant column to 'Permission'
khprj_df.rename(columns={'Folder Xây dựng KH PRJ': 'Permission'}, inplace=True)

# Keep only required columns
khprj_df = khprj_df[['Email', 'BU', 'Permission']]

# Optional: Display the first 50 rows for verification
# khprj_df.head(50)
khprj_df.count()
```

```python
# Folder RelativeURL templates for KH PRJ
khprj_folder_url_templates = {
    'Folder Xây dựng KH PRJ': '/sites/BSCKPI/Shared Documents/2025/{}/1. Xây dựng kế hoạch PRJ'
}

# Initialize an empty list to store the new rows
final_khprj_rows = []

# Iterate over each row in the khprj_df
for index, row in khprj_df.iterrows():
    # First, process the Permission value before URL generation
    processed_permission = map_permission(row['Permission'])

    # Generate new row based on the folder_url_template for KH PRJ
    for folder_name, folder_url_template in khprj_folder_url_templates.items():
        # Create a new row by copying the original one
        new_row = row.copy()

        # Generate the RelativeURL using the 'BU' column (or any other appropriate column)
        relative_url = folder_url_template.format(row['BU'])
        
        # Add the processed Permission and RelativeURL to the new row
        new_row['Permission'] = processed_permission
        new_row['RelativeURL'] = relative_url

        # Add the new row to the list
        final_khprj_rows.append(new_row)

# Convert the list of rows into a new DataFrame
khprj_df_extended = pd.DataFrame(final_khprj_rows)

# Save to Excel
filename = 'url_khprj.xlsx'
with pd.ExcelWriter(filename, engine='xlsxwriter') as writer:
    khprj_df_extended.to_excel(writer, index=False, sheet_name="data")

    # Access the xlsxwriter workbook and sheet
    workbook = writer.book
    worksheet = writer.sheets["data"]

    # Add a table format for the data
    worksheet.add_table(0, 0, khprj_df_extended.shape[0], khprj_df_extended.shape[1] - 1,
                        {'columns': [{'header': col} for col in khprj_df_extended.columns]})

    print(f'Saved {filename}')  # Optional, to confirm saving
```
