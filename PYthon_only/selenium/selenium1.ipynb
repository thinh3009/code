```python
import os
from selenium import webdriver
from selenium.webdriver.common.by import By   
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.keys import Keys#gui nut tren ban phim
import time
```

```python
driver=webdriver.Edge()
url='https://dosi-in.com/'
driver.get(url)
clickbutton=driver.find_element(By.XPATH,'//*[@id="dosiin_header"]/div[3]/div[1]/ul/li[2]/a')
driver.implicitly_wait(3)#cho cho den khi web load xong
clickbutton.click()

#second_element=driver.find_element(By.ID,'')

WebDriverWait(driver,4).until(
    EC.text_to_be_present_in_element(
        (By.XPATH,'//*[@id="dosiin_header"]/div[3]/div[1]/ul/li[2]/a'),#element filltration
        'complete!'#the expected text
    )
)
```
