```python
class Sol(object):
    def removeDuplicates(self, nums):
        unique_nums = []
        tong=0
        for num in nums:
            if num not in unique_nums:
                unique_nums.append(num)
                tong+=1
        return tong,unique_nums

answer = Sol()
k = answer.removeDuplicates([1, 1, 2,3,2,3,4,2,1])
print(k)  # Output: [1, 2]
```
