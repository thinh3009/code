```python
import time
print('put some skill that you not familar')
unfamilar_skill=input('>')
print(f'Filtering out {unfamilar_skill}')
def find_job():
    html_text= requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&searchTextSrc=ft&searchTextText=&txtKeywords=data+engineer&txtLocation=').text
    soup=BeautifulSoup(html_text,'lxml')
    job = soup.find_all('li',class_='clearfix job-bx wht-shd-bx')
    for index, j in enumerate(job):
        job_publish=j.find('span',class_='sim-posted').span.text
        if 'few' in job_publish:
            company_name=j.find('h3',class_='joblist-comp-name').text.replace(' ','')
            skills=j.find('span',class_='srp-skills').text.replace(' ','')
            more_infor=j.header.h2.a['href']
            if unfamilar_skill not in skills:#kiem tra xem ki nang nao khong co 
                with open(f'posts/{index}.txt','w') as f:

                    f.write(f'company name:{company_name.strip()} \n')
                    f.write(f'require skills:{skills.strip()}\n')
                    f.write(f'more info: {more_infor}\n')
                print('file saved: {index}')    

if __name__=="__main__":
    while True:
        find_job()
        time_wait=10
        print(f'waiting {time_wait} minute--')
        time.sleep(time_wait*60)
```
