```python
import requests
from bs4 import BeautifulSoup
html_text= requests.get('https://cafef.vn/').text
print(html_text)
```

```python
soup=BeautifulSoup(html_text,'lxml')
detail=soup.find_all('div',class_='tlitem box-category-item')
print(detail)
```

```python
for j in detail:
    title=j.find('h3')
    print(j)
```

```python
for j in detail:
    title=j.find('div',class_='knswli-right')
    title_text=title.h3.a
    print(title_text.text)
```

```python
from datetime import datetime
for j in detail:
    title=j.find('div',class_='knswli-right')
    title_text=title.h3.a.text#title
    category_all=j.find('p',class_='time_cate')
    category=category_all.find('a').text#loai thong tin
    time=j.find('span',class_='time time-ago')#thoi gian dang bai
    if time:
        time_string = time.text
        dt = datetime.strptime(time_string, "%Y-%m-%dT%H:%M:%S")
        formatted_time = dt.strftime("%d-%m-%Y %H:%M:%S")
    print(f"Name: {title_text}\
                \nCategory: {category}\
                \ntime: {formatted_time}\
                \n\n")
```
