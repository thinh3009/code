```python
from bs4 import BeautifulSoup
import requests
```

```python
htmltxt=requests.get('https://www.javdatabase.com/top-rated-jav-idols/i-all-time/').text
print(htmltxt)
```

```python
soup=BeautifulSoup(htmltxt,'lxml')
avlist=soup.find_all('div',class_="col-md-3 col-lg-2 col-xxl-2 col-4")
print(avlist)
```

```python
def find_idol1():
    for i in avlist:
        list_info=i.find_all('div',class_='card-body d-flex flex-column')
    
        
        for j in list_info: 
            link_idol=j.p.a['href']
            name_idol=j.find('a',class_='cut-text')   
            age_idol=j.find('a',class_='btn btn-primary btn-sm')
            cup_idol=j.find('a',class_='btn btn-info btn-sm')
            high_idol=j.find('a',class_='btn btn-secondary btn-sm')
            if age_idol and cup_idol and high_idol is not None:
                print(f"Name: {name_idol.text}\
                        \nLink: {link_idol}\
                        \nAge: {age_idol.text}\
                        \nCup: {cup_idol.text}\
                        \nHeigh: {high_idol.text}\n\n")
            else:
                print('NULL')
    
            
        
    

```

```python
import pandas as pd

data = []  # This will store the data
for i in avlist:
    list_info=i.find_all('div',class_='card-body d-flex flex-column')
    
    for j in list_info: 
        link_idol=j.p.a['href']
        name_idol=j.find('a',class_='cut-text')   
        age_idol=j.find('a',class_='btn btn-primary btn-sm')
        cup_idol=j.find('a',class_='btn btn-info btn-sm')
        high_idol=j.find('a',class_='btn btn-secondary btn-sm')

        if age_idol and cup_idol and high_idol is not None:
                # Create a dictionary for the row and append it to the data list
            data.append({
                'Name': name_idol.text,
                'Link': link_idol,
                'Age': age_idol.text,
                'Cup': cup_idol.text,
                'Height': high_idol.text
            })
        else:
            pass

# Create a DataFrame from the data
df = pd.DataFrame(data)
df
    
```
