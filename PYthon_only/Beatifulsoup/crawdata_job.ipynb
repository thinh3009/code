# Crawl data


```python
from bs4 import BeautifulSoup
import pandas as  pd
import requests
import time
import csv


def FindJob():
    input_not_familar=input('Put som skill not familar')
    print(f'Fillering out: {input_not_familar}')
    with open('job.csv','w',newline='') as file_out:
        header=['Position Name','Company','Skill require','More infor','Position and require','Time update','Location']
        writer=csv.DictWriter(file_out,delimiter=',',lineterminator='\n',fieldnames=header)
        writer.writeheader()

        
        htmltxt=requests.get('https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=python&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&pDate=I&sequence=5&startPage=1').text
        soup=BeautifulSoup(htmltxt,'html.parser')
        detail=soup.find_all('li',class_='clearfix job-bx wht-shd-bx')

        for i in detail:
            company=i.find('header',class_='clearfix')


            position_name=i.find('a')
        
            infor=i.find('ul',class_='top-jd-dtl mt-16 clearfix')
            location=infor.find('li', class_='srp-zindex location-tru')
            # print(location.text.strip().replace(', ',''))
            full_infor=','.join([list1.text.replace(',','').strip() for list1 in infor.find_all('li')])
        
            time_update=i.find('span',class_='sim-posted')

            another_skill=i.find('div',class_='more-skills-sections')
            skills = ','.join([span.text.strip() for span in another_skill.find_all('span')])
            more_infor=i.header.h2.a['href']
            comp_name=company.find('h3',class_='joblist-comp-name')
            if input_not_familar not in another_skill:
                print(f"Position Name: {position_name.text.strip()}")
                print(f"\nCompany: {comp_name.text.strip()}")
                print(f"\nSkill require: {skills}")
                    
                print(f"\nMore infor: {more_infor}")
                print(f"\nPosition and require: {full_infor}")
                print(f"\nTime update: {time_update.text.strip()}")
                print(f"\nLocation: {location.text.strip().replace(', ','')}")
                writer.writerow({
                    'Position Name':position_name.text.strip(),
                    'Company':comp_name.text.strip(),
                    'Skill require':skills,
                    'More infor':more_infor,
                    'Position and require':full_infor,
                    'Time update':time_update.text.strip(),
                    'Location':location.text.strip().replace(', ','')
                })
            else:
                print('Do not find any position')

    
```

# Clean data

```python
df_job=pd.read_csv('job.csv')

def CleanData():
    df_job[['loction1','Experience','Salary']]=df_job['Position and require'].str.split(',',expand=True)
    df_job.drop('loction1',axis=1,inplace=True)
    df_job.drop('Position and require',axis=1,inplace=True)
    df_job['Salary']=df_job['Salary'].str.replace(r'\s+', '', regex=True)
    df_job['Salary']=df_job['Salary'].str.replace('Notdisclosed','Not disclosed')
    df_job

```

# convert to excel

```python
def ConvertExcel():
    try:

        df_excel=df_job.to_excel("excel_job5.xlsx",sheet_name="sheet1",index=False)
        df_excel
    except:
        print("Error!")
    finally:
        print("Convert Finish!!!")       
ConvertExcel()
```

# run

```python
if __name__=="__main__":
    FindJob()
    CleanData()
    ConvertExcel()
```
