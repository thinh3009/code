```python
import pandas as pd
#dinh nghia kieu du lieu ngay thang cho cot day
df=pd.read_csv("C:\\Users\\buidu\\OneDrive\\Máy tính\\code2\\PYthon_only\\pandas\\data\\weather_data.csv",parse_dates=["day"])
df.set_index('day',inplace=True)
df
```

```python
new_df= df.fillna(0)#gan gia tri 0 cho cot k co du lieu
new_df
```

```python
#dat ten cho cac gia tri k co du lieu
new_df=df.fillna({
    'temperature':0,
    'winspeed':0,
    'event':'no event'
})
new_df
```

```python
#day cac gia tri tu o co du lieu xuong o ko co du lieu
new_df=df.fillna(method='ffill')
new_df
```

```python
#day cac gia tri tu o co du lieu len o ko co du lieu
new_df=df.fillna(method='bfill')
new_df
```

```python
#day cac gia tri tu o co du lieu sang trai o ko co du lieu
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

```python
#day cac gia tri tu o co du lieu xuong o ko co du lieu
#nhung chi day 1 gia tri duy nhat
new_df=df.fillna(method='bfill',limit=1)
new_df
```

```python
#dung thuat toan noi suy tuyen tinh de tinh gia tri (interpolate linear)
new_df=df.interpolate()
new_df
```

```python
#dung thuat toan noi suy tuyen tinh de tinh gia tri (interpolate linear)
#dua vao thoi gian
new_df=df.interpolate(method='time')
new_df
```

```python
#day nhung o co day du data
new_df=df.dropna()
new_df
```

```python
#day nhung gia tri co chua 2 tren tong 3 du lieu trong bang 
new_df=df.dropna(thresh=2)
new_df
```
