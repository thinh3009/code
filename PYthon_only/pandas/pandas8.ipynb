```python
import pandas as pd
df1=pd.DataFrame({
    'city':['newyork','chicago','orlando'],
    'temperature':[21,14,35]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['newyork','chicago','san francisco'],
    'humidity':[65,70,56]
})
df2
```

```python
#inner join cac data co cung du lieu
df3=pd.merge(df1,df2,on='city')
df3
```

```python
#outer join cac data co cung du lieu
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
#left join cac data co cung du lieu
df3=pd.merge(df1,df2,on='city',how='left')
df3
```

```python
#right join cac data co cung du lieu
df3=pd.merge(df1,df2,on='city',how='right')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

```python
df1=pd.DataFrame({
    'city':['new york','chicago','orlando','baltimorce'],
    'temperature':[21,22,34,56],
    'humidity':[65,56,67,78]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','chicago','san diago'],
    'temperature':[21,22,34],
    'humidity':[65,56,71]
})
df2
```

```python
#merge cac cot nhung khong co dieu kien join
#suffixes: thay doi thong tin o cac cot
df3=pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df3
```
