1. Tim kiem tuyen tinh

```python
import random
def sinhmang(n):
    mang=[]
    for i in range (n):
        so_ngauNhien=random.randint(-100,100)
        mang.append(so_ngauNhien)
    return mang
def timtuyentinh(mang,x):
    
    for i in range(len(mang)):
        if mang[i]==x:
            return i
    return -1
def main():
    mang=sinhmang(20)
    print(mang)

    x=int(input("nhap so muon tim: "))
    vitri=timtuyentinh(mang,x)
    if vitri!=-1:
        print(f"gia tri {x} duoc tim thay tai vi tri {vitri}")
    else:
        print("khong tim thay")
'''if __name__=="__main__":
    main()'''
```

2. tim kiem nhi phan

```python
import random

def sinhmangtangdan(n):
    mang=[]
    sodautien=random.randint(-100,100)
    mang.append(sodautien)  # You need to append the first random number to the list
    for i in range(1,n):
        tang=random.randint(1,10)
        so=mang[i-1]+tang
        mang.append(so)
    return mang

def timnhiphan(mang,x):
    trai=0
    phai=len(mang)-1
    while trai<=phai:
        giua=(trai+phai)//2
        if mang[giua]==x:  # You need to use mang[giua] instead of [giua]
            return giua
        if x< mang[giua]:
            phai=giua-1
        else:
            trai=giua+1
    return -1

def main():
    mang=sinhmangtangdan(20)
    print(mang)
    x=int(input("nhap gia tri can tim"))
    vitri=timnhiphan(mang,x)
    if vitri!=-1:
        print(f"tim thay phan tu {x} tai vi tri {vitri}")
    else:
        print("khong tim thay")

'''if __name__=="__main__":
    main()'''
```

4. sap xep chon

```python
import random
def sinhngaunhien(n):
    mang=[]
    for i in range(n):
        songaunhien=random.randint(1,100)
        mang.append(songaunhien)
    return mang
def sapxepchon(mang):
    n=len(mang)
    for i in range (n-1):
        vitriMin=i
        for j in range(i+1,n):
            if mang[vitriMin]>mang[j]:
                vitriMin=j
        mang[i],mang[vitriMin]=mang[vitriMin],mang[i]
        print(i+1,"-",mang)
def main():
    mang=sinhngaunhien(10)
    print('mang ban dau: ',mang)
    sapxepchon(mang)
    print("mang sau khi sap xep: ",mang)
if __name__=="__main__":
    main()
```
